```python
import cv2
import numpy as np

# load images
img1 = cv2.imread(r'./images/2.jpg')
w1 = img1.shape[1]

img2 = cv2.imread(r'./images/1.jpg')
w2 = img2.shape[1]

img3 = cv2.imread(r'./images/2.jpg')
w3 = img3.shape[1]

# get maximum width
ww = max(w1, w2, w3)
print(ww)
# pad images with transparency in width
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2BGRA)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2BGRA)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2BGRA)
print(ww-w2)
img1 = cv2.copyMakeBorder(img1, 0, 0, 0, 1131-1131, borderType=cv2.BORDER_CONSTANT, value=(0,0,0,0))
img2 = cv2.copyMakeBorder(img2, 0, 0, 0, 1131-w2, borderType=cv2.BORDER_CONSTANT, value=(0,0,0,0))
img3 = cv2.copyMakeBorder(img3, 0, 0, 0, 1131-w3, borderType=cv2.BORDER_CONSTANT, value=(0,0,0,0))
img4 = cv2.copyMakeBorder(img2, 0, 0, 0, 1131-w3, borderType=cv2.BORDER_CONSTANT, value=(0,0,0,0))

# stack images vertically
result = cv2.vconcat([img1, img2, img3,img4])

# write result to disk
cv2.imwrite("image_stack.png", result)

cv2.imshow("RESULT", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import random
from PIL import Image
import os
import glob

dataset_paths = [filename for filename in glob.glob(r'E:\Collage\images/*.jpg')]

for i, path in enumerate(dataset_paths):
    bg_im = Image.new(mode="RGB", size=(1131,1131), color="white")
    img = Image.open(path).resize((24, 24))
    
    bg_width, bg_height = bg_im.size
    img_width, img_height = img.size
    max_x = bg_width - (img_width * 2)
    max_y = bg_height - (img_height * 2)

    if max_x < 0 or max_y < 0:
        print("image cannot be completely pasted on background image")
        continue 

    random_x = random.randint(0, max_x)
    random_y = random.randint(0, max_y)
    print(random_x)
    print(random_y)
    bg_im.paste(img, box=(random_x, random_y))
    plt.imshow(bg_im)
#     bg_im.save(f'data/dst/{os.path.basename(path}')
```
